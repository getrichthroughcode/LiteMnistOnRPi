```python
import sys
import os
import scipy
import matplotlib.pyplot
import numpy
from PIL import Image
import IPython
import Xim_library
import numpy

pathDirectory="./imageXim/"
pathImage="2018-12-11 20-13-09-MV Triggered-6x/771429380/"

mallocMax=2731
completePath= numpy.array([ord(x) for x in list(pathDirectory+pathImage)],dtype='int32')
fileNumber = len(os.listdir(pathDirectory+pathImage))
lenPathImage = len(completePath)
```

```python

f = open(pathDirectory+pathImage+"00000.xim", "rb")
f.seek(12)
width=int.from_bytes(f.read(4),byteorder='little',signed=True)
height=int.from_bytes(f.read(4),byteorder='little',signed=True)
f.close()
```

```python

image = numpy.zeros(fileNumber*width*height,dtype='int32')
print("before: ", image)
for i in range((fileNumber//mallocMax)+1):
    if i!=(fileNumber//mallocMax):
        Xim_library.XimPyBind(image[i*mallocMax*width*height:(i+1)*mallocMax*width*height],mallocMax,completePath,lenPathImage, i, mallocMax)
    else:
        Xim_library.XimPyBind(image[i*mallocMax*width*height:i*mallocMax*width*height+(fileNumber-i*mallocMax)*width*height],fileNumber-i*mallocMax,completePath,lenPathImage, i, mallocMax)

print("after: ", image)
imageXim=image.reshape(fileNumber,height,width)
```

```python
matplotlib.pyplot.imshow(imageXim[200])
matplotlib.pyplot.gca().invert_yaxis()
matplotlib.pyplot.show(block=False)
```
